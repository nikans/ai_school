```python
data = pd.read_csv('abalone_data.csv')
```

```python
data.head()
```

```python
data.sex.unique()
```

```python
data['is_old'] = np.array([0 if r < 18 else 1 for r in data.num_rings.values])
```

```python
np.mean(data.is_old)
```

```python
import seaborn as sns
```

```python
sns.pairplot(data, hue='is_old')
```

```python
del data['num_rings']
```

```python
data['M'] = np.array([1 if s == 'M' else 0 for s in data['sex'].values])
```

```python
data['F'] = np.array([1 if s == 'F' else 0 for s in data['sex'].values])
```

```python
data['I'] = np.array([1 if s == 'I' else 0 for s in data['sex'].values])
```

```python
del data['sex']
```

```python
data.head()
```

```python
y = data['is_old']
```

```python
del data['is_old']
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(data, y, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC(random_state=42)
svc_model.fit(Xtrain, Ytrain)
```

```python
from sklearn.metrics import classification_report
```

```python
print( classification_report(Ytest, svc_model.predict(Xtest)) )
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sampler = SMOTE(random_state=42)
```

```python
Xres, Yres = sampler.fit_resample(Xtrain, Ytrain)
```
